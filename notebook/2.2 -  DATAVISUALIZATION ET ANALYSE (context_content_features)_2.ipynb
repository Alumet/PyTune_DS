### Imports

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

## Fichier context_content

```python
df_content = pd.read_csv('raw data/context_content_features.csv', error_bad_lines=False)
```

## Netoyage des données

```python
df_content = df_content.drop(columns=['coordinates', 'place', 'geo','user_id'])
```

```python
# Supression des données manquantes
df_content = df_content.dropna()
```

```python
# On prend un sous dataset pour des raison de manque de RAM
df_content = df_content.sample(frac=0.3, random_state=42)
```

### Distribution des événements d'écoute groupés par jour suivant l'heure de la journée

```python
# Extraction de l'heure
df_content['listening_h'] = df_content['created_at'].apply(lambda x: x[-8:-6]).astype('int')
```

```python
# Extraction du jour
df_content['listening_day'] = df_content['created_at'].apply(lambda x: x[5:10])
```

```python
# On regroupe les écoute par jours et on prend le compt
df_gb = df_content.groupby(["listening_day",'listening_h']).size().reset_index(name='count')
```

```python
ax = sns.catplot(data=df_gb, x='listening_h', y='count',kind = 'box', height=8.27, aspect=16/9)

plt.title("Distribution des événements d'écoute groupés par jour suivant l'heure de la journée")
ax.set(ylabel="Evénements d'écoute", xlabel='Heure de la journée', ylim=[200,550])
plt.savefig("Distribution des événements d'écoute groupés par jour suivant l'heure de la journée", dpi=300, pad_inches=0.1, bbox_inches='tight')
```

### Analyse d'indépendance entre le nombre découtes et l'heure de la journée

```python
import statsmodels.api

result = statsmodels.formula.api.ols(f'count ~ listening_h' ,data=df_gb).fit()
table = statsmodels.api.stats.anova_lm(result)
print(table)
```

### Analyse de l'influance du jour de la semaine sur les features

```python
from datetime import datetime

date = '2014-04-12 02:59:32'

def date_to_day(x):
    
    a={0: 'Monday',
       1: 'Tuesday',
       2: 'Wednesday',
       3: 'Thursday',
       4: 'Friday',
       5: 'Saturday',
       6: 'Sunday',
      }
    
    datetime_object = datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
    return a[datetime_object.weekday()]
    
date_to_day(date)
    
```

```python
df_content['Week_day'] = df_content['created_at'].apply(date_to_day)
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax = sns.boxplot(ax=ax, data=df_content, x='Week_day', y='danceability',
                order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])

plt.title("Influence du jour de la semaine sur la danceability de la musique")
ax.set(ylabel="danceability", xlabel='Jour de la semaine')
plt.savefig("Influence du jour de la semaine sur la danceability de la musique", dpi=300, pad_inches=0.1,  bbox_inches='tight')
```
